```python
#Question 3

class Marsupial:
        
        def __init__(self,items):
            self.items = items
            
        def put_in_pouch(self,x):
            self.items.append(x)

        def pouch_contents(self):
            print(self.items)

class Kangaroo(Marsupial):

        def __init__(self, items, x, y):
            self.x = x
            self.y = y
            super().__init__(items)
            

        def setx(self, x):
            self.x = x

        def sety(self, y):
            self.y = y

        def get(self):
            return (self.x, self.y)

        def jump(self, dx, dy):
            self.x += dx
            self.y += dy
    
        def Print(self):
            print("I am a Kangaroo located at Coordinates (", self.x,",",self.y,")")

        
m = Marsupial([])
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()

k = Kangaroo([],0,0)
k.Print ()
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
k.Print()
    
```

```python
#Question 6

from html.parser import HTMLParser                  
class HeadingParser(HTMLParser):
    header=False
    header2 = False 
    def handle_starttag(self,tag,attrs):
        if tag=="h1":
            self.header =True
        elif tag =="h2":
            self.header2 = True
    
            
    def handle_data(self, data):
        if self.header:
            print(data)
        elif self.header2:
            print('\t','\t',data)
        
            
hp = HeadingParser()
infile = open('w3c.html')
content = infile.read()
infile.close()
hp.feed(content)
```
